```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```
# File location and type
file_location = "/FileStore/tables/wikiaves_metadata_2019_08_11.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
data = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(data)

data.createOrReplaceTempView("data")
```

```
data.count()
```

```
sql="""
    select last_day(registry_date) as mon
    ,count(*) as total_records
    ,count(distinct author_id) as authors_count
    ,count(distinct species_id) as unique_species_count
    from data
    group by last_day(registry_date)
    order by last_day(registry_date)

"""

monthly_stats=spark.sql(sql).toPandas()


monthly_stats["mon"]=pd.to_datetime(monthly_stats["mon"])
display(monthly_stats)
```

```
# Monthly total records plot
plt.figure(figsize=(12,6))
plt.plot(monthly_stats["mon"],monthly_stats["total_records"],marker="o")
plt.title("monthly total registry records",fontsize=14)
plt.show()
```

```
# Comparion between UNIQUE species count found monthly and total UNIQUE authors count monthly plot
plt.figure(figsize=(12,6))
plt.plot(monthly_stats["mon"],monthly_stats["authors_count"],marker="o",label="authors_count")
plt.plot(monthly_stats["mon"],monthly_stats["unique_species_count"],marker="*",label="unique_species_count")
plt.title("monthly total registry authors & species",fontsize=14)
plt.legend()
plt.show()
```

```
# Extract the UNIQUE species count and location/year data for location analysis
sql="""
    select a.location_name,year(registry_date) as year
    ,count(distinct species_id) as unique_species_count
    from data a 
    join
    (
        select location_name,total,row_number()over(order by total desc) as rn
        from
        (
            select location_name,count(distinct species_id) as total
            from data
            group by location_name
        )tmp
    )b
    on a.location_name=b.location_name
    and b.rn<=20
    group by a.location_name,year(registry_date)
    order by a.location_name,year(registry_date)

"""

location_yearly_stats=spark.sql(sql).toPandas()
location_yearly_stats.head()
```

```
# Unique bird species discovery counts plotting by location and year
location_yearly_pivot=location_yearly_stats.pivot_table(index='location_name',
                                                        columns='year',
                                                        values="unique_species_count",
                                                        aggfunc="sum",
                                                        margins=True)
location_yearly_pivot=location_yearly_pivot.iloc[:-1,:].sort_values(by="All",ascending=False).iloc[:,:-1].fillna(0).astype("int")
plt.figure(figsize=(20,10))

sns.heatmap(location_yearly_pivot,
            cmap="YlGnBu",
            cbar=True,
            annot=True, 
            linewidths=.25,
            fmt="d")
```

```
sql="""
    select 'top20' as category
    ,a.scientific_species_name
    ,case when year(registry_date)<2017 then 'before 2017' else 'after 2017' end as ind
    ,count(*) as total_records
    ,count(distinct location_id) as locations_count
    from data a 
    join
    (
        select scientific_species_name,total,row_number()over(order by total desc) as rn
        from
        (
            select scientific_species_name,count(*) as total
            from data
            group by scientific_species_name
        )tmp
    )b
    on a.scientific_species_name=b.scientific_species_name
    and b.rn<=20    
    where year(registry_date) in(2015,2016,2017,2018)
    group by 1,2,3
    
    union
    select 'bottom20' as category
    ,a.scientific_species_name
    ,case when year(registry_date)<2017 then 'before 2017' else 'after 2017' end as ind
    ,count(*) as total_records
    ,count(distinct location_id) as locations_count
    from data a 
    join
    (
        select scientific_species_name,total,row_number()over(order by total) as rn
        from
        (
            select scientific_species_name,count(*) as total
            from data
            group by scientific_species_name
        )tmp
    )b
    on a.scientific_species_name=b.scientific_species_name
    and b.rn<=20    
    where year(registry_date) in(2015,2016,2017,2018)
    group by 1,2,3


"""

specy_change_stats=spark.sql(sql).toPandas()
specy_change_stats
```

```
plt.figure(figsize=(9,6))
sns.barplot(x="scientific_species_name",y="total_records",hue="ind",data=specy_change_stats[specy_change_stats.category=="top20"])
plt.xticks(rotation=-90)
plt.title("top 20 species' records count before vs after 2017",fontsize=14)
plt.show()
```

```
plt.figure(figsize=(9,6))
sns.barplot(x="scientific_species_name",y="total_records",hue="ind",data=specy_change_stats[specy_change_stats.category=="bottom20"])
plt.xticks(rotation=-90)
plt.title("bottom 20 species' records count before vs after 2017",fontsize=14)
plt.show()
```

```
import logging
from fbprophet import Prophet
 
logger = spark._jvm.org.apache.log4j
logging.getLogger("py4j").setLevel(logging.ERROR)
```

```
%sql
DROP TABLE IF EXISTS birds;
 
CREATE TABLE birds
USING csv
OPTIONS (path "/FileStore/tables/wikiaves_metadata_2019_08_11.csv", header "true")
```

```
# Extract registry date and total registries data
query = """
 SELECT
    string(registry_date) as ds,
    COUNT(registry_id) as y
FROM
    `birds`
GROUP BY ds
ORDER BY ds;
"""
 
date = spark.sql(query)
time = date.toPandas()
display(time)
 
```

```
# logarithm of predict target
import numpy as np
time['y'] = np.log(time['y'])
display(time)
```

```
# Model building and predict
m = Prophet(daily_seasonality=True)
 
m.fit(time)
 
future = m.make_future_dataframe(periods=730)
 
forecast = m.predict(future)
```

```
# Result plotting
from fbprophet.plot import add_changepoints_to_plot
fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), m, forecast)
```

```
# Check the predictions
future.tail()
```

```
# Cyclic trend plotting
m.plot_components(forecast)
```

```
# Check about features of the time series model
print(forecast.columns)
```

```

from fbprophet.diagnostics import cross_validation
df_cv = cross_validation(m, initial='730 days', period='180 days', horizon = '365 days')
```

```
df_cv.tail()
```

```
from fbprophet.diagnostics import performance_metrics
df_p = performance_metrics(df_cv)
df_p.tail()
```

```
# Plot the metric result 
from fbprophet.plot import plot_cross_validation_metric
fig = plot_cross_validation_metric(df_cv, metric='mape')
plt.ylim([0,0.125])
```

```

query = """
SELECT
     is_large,is_flagged,like_count,views_count,species_id
    
FROM
    `birds`
"""
 
df = spark.sql(query)
data = df.toPandas()
```

```
# Unify the data types
data["is_large"] = data["is_large"].apply(float)
data["is_flagged"] = data["is_flagged"].apply(float)
data["like_count"] = data["like_count"].apply(float)
data["views_count"] = data["views_count"].apply(float)
data["species_id"] = data["species_id"].apply(float)
```

```
data = data[["is_large","is_flagged","like_count","views_count","species_id"]]
```

```
# Prepare train and test dataset
from sklearn.model_selection import train_test_split
train , test = train_test_split(data, test_size = 0.01)
train = train.dropna()
test = test.dropna()
 
x_train = train.drop(["species_id"], axis = 1)
x_test = test.drop(["species_id"], axis = 1)
y_train = train["species_id"]
y_test = test["species_id"]
```

```
from sklearn import neighbors
import math
from sklearn.metrics import mean_squared_error
model = neighbors.KNeighborsRegressor(n_neighbors = 3)
model.fit(x_train, y_train)
```

```

model = neighbors.KNeighborsRegressor(n_neighbors = 2)
model.fit(x_train, y_train)
 
#RMSE
pred=model.predict(x_test)
error = math.sqrt(mean_squared_error(y_test,pred))
print(error)
```

```
from sklearn import neighbors
import math
from sklearn.metrics import mean_squared_error
model = neighbors.KNeighborsRegressor(n_neighbors = 3)
model.fit(x_train, y_train)
 
#RMSE
pred=model.predict(x_test)
error = math.sqrt(mean_squared_error(y_test,pred))
print(error)
```

```
# Prediction accuracy check
predict = model.predict(x_test)
final = x_test.copy()
final['predict_Sepcies'] = predict
final_price = y_test
final["species_id"] = final_price
print("Test set accuracy: {:.2f}".format(model.score(x_test, y_test)))
```
